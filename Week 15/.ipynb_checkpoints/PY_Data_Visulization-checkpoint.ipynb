```python
fig, ax = plt.subplots()

# Plotting the cumulative sum of random data on the specified axes
ax.plot(np.random.randn(1000).cumsum())

plt.show()
```

```python
# Setting custom x-axis ticks
ax.set_xticks([0, 250, 500, 750, 1000])

plt.show()
```

```python
ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

# Setting labels for the x-axis and y-axis
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

# Setting the title of the plot
ax.set_title('My first Python graph')

plt.show()
```

```python
ax
```

```python
fig
```

```python
# Dictionary containing title, xlabel, and ylabel
props = {'title': "My first Graph",
         'xlabel': 'Stages',
         'ylabel': 'Values'}

# Setting properties using the dictionary
ax.set(**props)

fig
```
