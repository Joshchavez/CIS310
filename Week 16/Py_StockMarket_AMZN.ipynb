```python
from pandas.api.types import is_numeric_dtype

numeric_count = 0
non_numeric_count = 0

for c in data.columns:
    if is_numeric_dtype(data[c]):
        numeric_count += 1
    else:
        non_numeric_count += 1

print(f"Numerical columns: {numeric_count}")
print(f"Non-numeric columns: {non_numeric_count}")
```

```python
#remove all columns that have a constant/single value 
constant_columns = []

for c in data.columns:
    n = len(data[c].unique())
    if n == 1:
        constant_columns.append(c)
        
print(f"Constant columns are: {constant_columns}")
```
